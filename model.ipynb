```python
import pandas as pd
import numpy as np
import os
import sqlite3
```

```python
DB_NAME='Electricity.db'
DB_PATH=os.path.join(os.getcwd(),DB_NAME)

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()
```

```python
main_data_from_db=cur.execute("SELECT * FROM main_data")
col_name=[column[0] for column in main_data_from_db.description]
main_df=pd.DataFrame.from_records(data=main_data_from_db.fetchall(), columns=col_name)

total_data_from_db=cur.execute("SELECT * FROM total_data")
col_name_tot=[column[0] for column in total_data_from_db.description]
total_df=pd.DataFrame.from_records(data=total_data_from_db.fetchall(), columns=col_name_tot)

main_df=main_df.replace('광업 ','광업')
main_df=main_df.replace(['제조업 ','금융 및 보험업 ','농업, 임업 및 어업 ','숙박 및 음식점업 ','전기, 가스, 증기 및 수도사업 ','전문, 과학 및 기술 서비스업 ','건설업 ','부동산업 및 임대업 '],
['제조업','금융 및 보험업','농업, 임업 및 어업','숙박 및 음식점업','전기, 가스, 증기 및 수도사업','전문, 과학 및 기술 서비스업','건설업','부동산업 및 임대업'])

total_df=total_df.replace('광업 ','광업')
total_df=total_df.replace(['제조업 ','금융 및 보험업 ','농업, 임업 및 어업 ','숙박 및 음식점업 ','전기, 가스, 증기 및 수도사업 ','전문, 과학 및 기술 서비스업 ','건설업 ','부동산업 및 임대업 '],
['제조업','금융 및 보험업','농업, 임업 및 어업','숙박 및 음식점업','전기, 가스, 증기 및 수도사업','전문, 과학 및 기술 서비스업','건설업','부동산업 및 임대업'])

main_df.to_csv('main_df.csv',index=False)
total_df.to_csv('total_df.csv',index=False)

conn.close()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline

target='bill_per_cust'
# feature=['year','month','metro','city','biz','power_per_cust']
feature=['year','metro','city','biz','power_per_cust']
main_df=main_df.dropna()
x = main_df[feature]
y = main_df[target]


x_train, x_test, y_train, y_test = train_test_split(x , y, test_size=0.2, random_state=99)
# x_train, x_val , y_train, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=99)

```

```python
from sklearn.metrics import confusion_matrix
from sklearn.compose import ColumnTransformer

pipe = make_pipeline(
    # ColumnTransformer([('one_hot_encoder', OneHotEncoder(),['month','metro','city','biz'])],remainder='passthrough'),
    ColumnTransformer([('one_hot_encoder', OneHotEncoder(),['metro','city','biz'])],remainder='passthrough'),
    # OneHotEncoder(categories=['month','metro','city','biz']),
    XGBRegressor()
)

pipe.fit(x_train,y_train)
# encoder = OneHotEncoder()
# encoder.fit_transform(x_train)
y_pred_test = pipe.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred_test)
```
